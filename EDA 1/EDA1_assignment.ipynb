```python
# Function to detect outliers
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)

# List of numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Detect outliers in numerical columns
outliers = pd.DataFrame()
for col in numerical_columns:
    outliers[col] = detect_outliers_iqr(df[col])

# Treat outliers by capping them to the 5th and 95th percentiles
for col in numerical_columns:
    lower_bound = df[col].quantile(0.05)
    upper_bound = df[col].quantile(0.95)
    df[col] = df[col].clip(lower_bound, upper_bound)
```

```python
# Creating Histograms for Numerical Columns

import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical columns
numerical_columns = df_filled.select_dtypes(include=['float64', 'int64']).columns

# Calculate the number of rows and columns required for the subplots
num_columns = len(numerical_columns)
num_rows = (num_columns + 2) // 3  # Divide by 3 columns, add 2 to round up

# Create histograms for all numerical columns
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(num_rows, 3, i)  # Adjust grid based on the number of columns
    sns.histplot(df_filled[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
#Summary of histograms of numerical columns
#Baseline Fetal Heart Rate (LB): The distribution is fairly normal, centered around 130-140 bpm, with slight variations on both sides. This indicates that most fetuses have a baseline heart rate within this range, suggesting normal fetal health.

#Accelerations (AC), Fetal Movements (FM), Uterine Contractions (UC), and Decelerations (DL, DS, DP): These variables show heavily skewed distributions towards zero, indicating that these events are relatively rare. Most values are either close to or exactly zero.

#Percentage of Time with Abnormal Short-Term Variability (ASTV) and Percentage of Time with Abnormal Long-Term Variability (ALTV): These variables show a broad range of values, with many cases having lower values but a significant tail towards higher values. This suggests that some fetuses exhibit prolonged periods of abnormal variability, which might warrant closer monitoring.

#Mean Value of Short-Term Variability (MSTV) and Mean Value of Long-Term Variability (MLTV): The distribution is fairly symmetrical, showing consistency in the variability of fetal heart rate over both short and long periods for most observations.
```
