```python
df = df.drop_duplicates()
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Impute missing values for numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute missing values for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

# Check for missing values again
print(df.isnull().sum())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standard Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])
df_scaled = pd.DataFrame(scaled_data, columns=numerical_columns)

# Min-Max Scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])
df_minmax = pd.DataFrame(scaled_data, columns=numerical_columns)
print('scaling techniques applyed sucessfully')
```

```python
#When to Use Each Scaling Technique
#Standard Scaling is preferred when:
#The data follows a normal (Gaussian) distribution.

#You are using algorithms that assume the data is normally distributed (e.g., linear regression, logistic regression).

#You need to handle outliers carefully, as it is less sensitive to outliers.

#Min-Max Scaling is preferred when:
#The data does not follow a normal distribution.

#You want to maintain the relationships between features.

#You are using algorithms that do not assume any specific distribution of the data (e.g., k-nearest neighbors, neural networks).
```

```python
#Encoding Techniques:
#Apply One-Hot Encoding to categorical variables with less than 5 categories.
#Use Label Encoding for categorical variables with more than 5 categories.
#Discuss the pros and cons of One-Hot Encoding and Label Encoding.
```

```python
#Applying one-hot encoding to categorical columns
import pandas as pd
for col in categorical_columns:
    if df[col].nunique() < 5:
        df_dummies = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, df_dummies], axis=1)
        df.drop(col, axis=1, inplace=True)

# Print the updated DataFrame
print(df.head())
```

```python
# applying Label encoding for categorical variables with more than 5 categories
from sklearn.preprocessing import LabelEncoder

categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    if df[col].nunique() > 5:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])

# Print the updated DataFrame
print(df.head())
```

```python
#Pros and Cons of One-Hot Encoding and Label Encoding
#One-Hot Encoding
#Pros:
#No Assumptions About Order: One-Hot Encoding does not assume any ordinal relationship among the categories, making it suitable for nominal categorical variables.

#Simplicity: Easily interpretable, as each category is represented by a separate binary column.

#Model Compatibility: Works well with most machine learning algorithms, especially those that can handle sparse data (e.g., tree-based methods).

#Cons:
#High Dimensionality: Can lead to a large increase in the number of features, especially for categorical variables with many unique values. This may cause computational inefficiencies and overfitting.

#Sparsity: Results in sparse matrices, which may not be ideal for algorithms that cannot handle sparse data well (e.g., linear regression).
```

```python
#Label Encoding
#Pros:
#Compact Representation: Encodes the categories into integer values, which is more compact than One-Hot Encoding.

#Low Dimensionality: Does not increase the number of features, making it more efficient in terms of memory and computation.

#Cons:
#Assumption of Order: Implies an ordinal relationship among categories, which may not be appropriate for nominal categorical variables.

#Interpretability: Can be less interpretable, as the numerical labels do not have any intrinsic meaning.
```

```python
# Create Education Level Group feature
education_levels = {
    'Bachelors': 'Higher',
    'Masters': 'Higher',
    'Doctorate': 'Higher',
    'Assoc-acdm': 'Intermediate',
    'Assoc-voc': 'Intermediate',
    'Some-college': 'Intermediate',
    '12th': 'Intermediate',
    '10th': 'Basic',
    '11th': 'Basic',
    '9th': 'Basic',
    '7th-8th': 'Basic',
    '5th-6th': 'Basic',
    '1st-4th': 'Basic',
    'Preschool': 'Basic'
}

df['education_level_group'] = df['education'].map(education_levels)
#Rationale: Grouping education levels into broader categories can simplify the data and help the model capture educational attainment better.
```

```python
# Create Hours Worked per Day feature
df['hours_per_day'] = df['hours_per_week'] / 7
#Rationale: Calculating the average number of hours worked per day can help the model understand work-life balance better.
```

```python
#Transformation of Skewed Numerical Feature
#Let's say we have a feature like "capital_loss" which might be skewed. We can apply a square root transformation to normalize its distribution.

#Rationale: Square root transformation can help in stabilizing the variance and making the data more normally distributed, which can improve the performance of some models

# Apply square root transformation to capital_loss
df['sqrt_capital_loss'] = df['capital_loss'].apply(lambda x: np.sqrt(x))
```

```python
#Feature Selection:
#Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance.
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
# Select numerical features for outlier detection
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = iso_forest.fit_predict(df[numerical_features])

# Add outlier column
df['outlier'] = outliers

# Filter out the outliers
df_cleaned = df[df['outlier'] != -1]

# Drop the outlier column
df_cleaned = df_cleaned.drop(columns=['outlier'])
```

```python
# Compute Pearson correlation coefficient manually
correlation_matrix = df_cleaned[numerical_features].corr()
pps_matrix = correlation_matrix.applymap(lambda x: np.square(abs(x)))
```
